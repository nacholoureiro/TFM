## Sesión 2- Números aleatorios y estimación del número $\pi$ - Resuelto

En esta sesión, intentaremos estimar el valor de $\pi$.\
Para ello, "simularemos" dardos disparados aleatoriamente a una diana circular de radio *r* como la de la imagen inscrita en un cuadrado de lado *2r*.



Según la Regla de Laplace, la probabilidad de que un dardo lanzado aleatoriamente caiga dentro del círculo, $P(dentro)$, será:

$P(dentro)=\frac{N_{dentro}}{N_{total}}$

donde $N_{dentro}$ es el número de dardos que caen dentro del círculo y $N_{total}$ es el número total de dardos lanzados.


Esta probabilidad puede expresarse como la razón entre las áreas del círculo y el cuadrado:

$P(dentro)=\frac{N_{dentro}}{N_{total}}=\frac{Área_{círculo}}{Área_{cuadrado}}=\frac{\pi r^{2}}{(2r)^{2}}=\frac{\pi r^{2}}{4 r^{2}}=\frac{\pi}{4}$

De aquí, obtenemos que

$\pi=4\frac{N_{dentro}}{N_{total}}$

## Solución (no es la única, ¡hay muchas formas de hacerlo!)


Vamos a suponer que nuestro círculo está centrado en el origen de coordenadas O (0,0) y tiene un radio $r$=1.\
Nuestro cuadrado, por lo tanto, tendrá $l=2r=2$. 

Un punto P(x,y) pertenecerá al círculo si se cumple que: 

$x²+y²<=1$

De esta manera, se contabilizan tanto los puntos que están dentro del círculo como aquellos situados sobre la circunferencia.

```python
# Importamos las librerías random y numpy. La librería matplotlib.pyplot sirve para hacer gráficos.
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos una función que tome como argumento únicamente el número total de dardos lanzados.
def pi_aprox(N):
    
    # Generamos de manera aleatoria N puntos (x,y) dentro del CUADRADO, es decir, en el intervalo [-1,1]
    x=np.random.uniform(-1, 1, size=(N)) 
    y=np.random.uniform(-1, 1, size=(N)) 
    
    # Los puntos que estén dentro de o sobre el CÍRCULO de radio 1 serán "interiores". Los demás serán "exteriores".
    interior = (x**2 + y**2) <= 1 
    exterior = np.invert(interior) 
    
    # Se estima el valor de PI y el % de error
    pi_est = interior.sum() * 4 / N 
    error = (abs(pi_est - np.pi)/np.pi)*100
  
    # Representamos gráficamente. Azul para puntos interiores, rojo para puntos exteriores
    plt.figure(figsize=(7,7))  
    plt.plot(x[interior], y[interior], 'b.')  
    plt.plot(x[exterior], y[exterior], 'rx') 

    print('Dardos lanzados:',N)
    print('Dardos dentro del círculo:',interior.sum())
    print('El valor real de Pi es',np.pi)
    print('El valor estimado de Pi es',pi_est)
    print('Esto supone un error del',round(error,3),'%')  
```
