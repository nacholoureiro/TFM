## Sesión 6- Máquina de Galton - Resuelto

En esta sexta y última sesión se pretende demostrar experimentalmente el *Teorema del Límite Central*. Este teorema afirma que, para una muestra aleatoria lo suficientemente grande (*N tiende a infinito*), la distribución de las medias sigue una distribución normal.  Con el objetivo de demostrar dicho teorema, se trabajará con *la Máquina de Galton*. Esta consiste en un tablero vertical con clavos uniformemente espaciados clavados en su mitad superior -en la que los clavos están dispuestos en orden escalonado- y una mitad inferior dividida en varias ranuras rectangulares espaciadas uniformemente. En el medio de la mitad superior hay un embudo por el que se dejan caer bolas. Cada vez que una bola golpea uno de los clavos, puede rebotar hacia la derecha con probabilidad **p** o hacia la izquierda con probabilidad **q**. Para clavos colocados simétricamente, las bolas rebotarán hacia la izquierda o hacia la derecha con la misma probabilidad **p** = **q** = 0.5. Este proceso se repite una y otra vez hasta que la bola termina en una cierta posición y cae por una de las ranuras de la mitad inferior. Si el número de bolas lanzado es lo suficientemente grande, se podrá observar cómo la distribución binomial de partida tiende a la distribución normal (campana de Gauss).


## Solución (no es la única, ¡hay muchas formas de hacerlo!)


```python
import random

def galton(altura, bolas):
    print("Máquina de Galton con",altura,"pisos,",altura+1,"ranuras y",bolas,"bolas.")
    # Creamos una lista vacía con las ranuras.
    # Hay una ranura más que alturas tiene la máquina de Galton.    
    
    ranuras = [0] * (altura+1) 
    
    #Simulamos la caída de la bola
    for i in range(bolas):
        pos = 0
        for j in range(0, altura):
            #En cada piso, la bola puede rebotar hacia la izquierda (0) o hacia la derecha (1)
            turn = random.randint(0, 1) 
            if turn == 1:
                pos += 1
        ranuras[pos] = ranuras[pos] + 1 # Sumamos una unidad a la ranura en la que ha caído la bola
    print("Se muestra, a continuación, cuántas bolas han caído en cada ranura\n",ranuras)
    return ranuras



def grafica(dataset):   
     # Representamos gráficamente y observaremos cuántas bolas han caído en cada ranura
     # Cada bola viene representada por un .
    print("REPRESENTACIÓN GRÁFICA DE LA SIMULACIÓN ANTERIOR\n")
    for val in dataset:
        for i in range(0, val):
            print('.', end='')
        print(val)
    print("\n¡AHÍ ESTÁ LA CAMPANA DE GAUSS!")

```

```python
ranuras=galton(altura=15, bolas=400)
```
